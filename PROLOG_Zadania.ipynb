W zadaniu użyłem:

* findall do zebrania wszystkich możliwych tras i kosztów
* sort do posortowania tras według kosztu
* not do sprawdzania, czy miasto nie jest już w trasie
* append do rozbudowywania listy trasy
* is do obliczeń arytmetycznych, np. sumowania kosztów
* member do sprawdzania obecności miasta w trasie

**wyniki zapytań:**

polaczenie(warszawa, gdansk).
true

koszt_podrozy(warszawa, gdansk, C).
C = 500

najkrotsza_droga(warszawa, gdansk).
Trasa: [warszawa,krakow,gdansk], Koszt: 500

```python
%Fakty
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).

polaczenie(warszawa, krakow, 300).
polaczenie(krakow, warszawa, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(gdansk, krakow, 200).
polaczenie(gdansk, warszawa, 500).

%Reguły
polaczenie(X, Y) :- polaczenie(X, Y, _).

koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).

najkrotsza_droga(Poczatek, Cel) :-
    findall((Koszt, Trasa), koszt_podrozy_wszystkie(Poczatek, Cel, 0, [Poczatek], Koszt, Trasa), Trasy),
    sort(Trasy, [(NajmniejszyKoszt, NajkrotszaTrasa)|_]),
    format('Trasa: ~w, Koszt: ~w~n', [NajkrotszaTrasa, NajmniejszyKoszt]).

koszt_podrozy_wszystkie(X, Y, AkumulowanyKoszt, AkumulowanaTrasa, KosztKoncowy, TrasaKoncowa) :-
    X == Y,
    KosztKoncowy = AkumulowanyKoszt,
    TrasaKoncowa = AkumulowanaTrasa.

koszt_podrozy_wszystkie(X, Y, AkumulowanyKoszt, AkumulowanaTrasa, KosztKoncowy, TrasaKoncowa) :-
    polaczenie(X, Z, Koszt),
    not(member(Z, AkumulowanaTrasa)),
    NowyAkumulowanyKoszt is AkumulowanyKoszt + Koszt,
    append(AkumulowanaTrasa, [Z], NowaAkumulowanaTrasa),
    koszt_podrozy_wszystkie(Z, Y, NowyAkumulowanyKoszt, NowaAkumulowanaTrasa, KosztKoncowy, TrasaKoncowa).
```

Zadanie polega na zarządzaniu stanami magazynowymi produktów: sprawdzaniu dostępności, przenoszeniu między sektorami oraz uzupełnianiu zapasów. Dodałem więcej faktów, aby sprawdzić poprawne działanie reguł oraz symulowanie prawdziwego magazynu.

W zadaniu użyłem:

* dynamic do dynamicznego definiowania faktów, które można zmieniać w trakcie działania programu
* retract do usuwania istniejących faktów
* assertz do dodawania nowych faktów na końcu bazy danych
* format do wyświetlania zmienionych danych w czytelnej formie
* is do obliczeń arytmetycznych, np. aktualizacji ilości produktów

**wyniki zapytań:**

dostepny_produkt(tv).
true

przenies_produkt(tv, sektor_b).
produkt(tv, 10, sektor_b)

uzupelnij_stan(pralka, 3).
produkt(pralka, 8, sektor_b)

```python
%fakty
:- dynamic produkt/3.

produkt(tv, 10, sektor_a).
produkt(pralka, 5, sektor_b).
produkt(lodowka, 3, sektor_c).
produkt(zmywarka, 7, sektor_d).
produkt(mikrofalowka, 4, sektor_e).
produkt(komputer, 15, sektor_a).
produkt(telefon, 20, sektor_a).
produkt(tablet, 12, sektor_a).
produkt(drukarka, 6, sektor_a).
produkt(odkurzacz, 8, sektor_b).

%reguły
dostepny_produkt(X) :-
    produkt(X, Ilosc, _),
    Ilosc > 0.

przenies_produkt(X, Y) :-
    produkt(X, Ilosc, StarySektor),
    retract(produkt(X, Ilosc, StarySektor)),
    assertz(produkt(X, Ilosc, Y)),
    format('produkt(~w, ~w, ~w)', [X, Ilosc, Y]).

uzupelnij_stan(X, N) :-
    produkt(X, Ilosc, Sektor),
    NowaIlosc is Ilosc + N,
    retract(produkt(X, Ilosc, Sektor)),
    assertz(produkt(X, NowaIlosc, Sektor)),
    format('produkt(~w, ~w, ~w)', [X, NowaIlosc, Sektor]).
```

# **ZADANIE 4**

Zadanie polega na wyszukiwaniu tras między miastami, obliczaniu czasu przejazdu oraz wskazywaniu najkrótszej trasy z podaniem pierwszego miasta na trasie i całkowitego czasu. W zadaniu dodałem też dwa nowe miasta.

W zadaniu użyłem:

* findall do zebrania wszystkich możliwych tras i ich czasów przejazdu.
* sort do posortowania tras według czasu przejazdu w celu znalezienia najkrótszej.
* member do sprawdzania, czy miasto zostało już odwiedzone, aby unikać cykli.
* \+/ (negacja) do sprawdzania, czy miasto nie jest na liście odwiedzonych.
* is do obliczeń arytmetycznych, np. sumowania czasu przejazdu.
* format do wyświetlania wyników w czytelnej formie.

**wyniki zapytań:**

droga(warszawa, krakow).
true

czas_przejazdu(warszawa, poznan, T).
T = 7

najkrotsza_trasa(warszawa, poznan).
Trasa przez krakow, czas 7

```python
% Fakty
miasto(warszawa).
miasto(krakow).
miasto(poznan).
miasto(lublin).
miasto(wroclaw).

droga(warszawa, krakow, 4).
droga(warszawa, poznan, 7).
droga(warszawa, lublin, 2).
droga(poznan, wroclaw, 3).
droga(poznan, warszawa, 7).
droga(krakow, poznan, 3).
droga(wroclaw, poznan, 3).

droga(X, Y) :- droga(X, Y, _).

czas_przejazdu(X, Y, T) :-
    czas_przejazdu_helper(X, Y, T, _, [X]).

czas_przejazdu_helper(X, Y, T, Trasa, Odwiedzone) :-
    droga(X, Y, T),
    \+ member(Y, Odwiedzone),
    Trasa = [X,Y].

czas_przejazdu_helper(X, Y, T, Trasa, Odwiedzone) :-
    droga(X, Z, T1),
    \+ member(Z, Odwiedzone),
    czas_przejazdu_helper(Z, Y, T2, ResztaTrasy, [Z|Odwiedzone]),
    T is T1 + T2,
    Trasa = [X|ResztaTrasy].

znajdz_najkrotsza_trase(X, Y, PierwszeMiasto, Czas) :-
    findall((Czas, Trasa), czas_przejazdu_helper(X, Y, Czas, Trasa, [X]), Pary),
    sort(Pary, PosortowanePary),
    PosortowanePary = [(MinCzas, MinTrasa)|_],
    Czas = MinCzas,
    MinTrasa = [_ ,PierwszeMiasto|_].

najkrotsza_trasa(X,Y) :-
    znajdz_najkrotsza_trase(X, Y, PierwszeMiasto, Czas),
    format('Trasa przez ~w, czas ~w', [PierwszeMiasto, Czas]).
```

Użyłem:

* findall/3 do zebrania listy książek zgodnych z gatunkiem i ocenionych pozytywnie przez użytkownika
* writeln/1 do wyświetlania listy polecanych książek w czytelnej formie
* dynamic/1 do umożliwienia dynamicznego zarządzania ocenami książek
* *>*=/2 do filtrowania książek z oceną równą lub wyższą od progu

**wyniki zapytań:**

polecane_ksiazki(uzytkownik1, fantasy).
[Hobbit, Gra o tron]

polecane_ksiazki(uzytkownik2, fantasy).
[Harry Potter i Kamień Filozoficzny]

polecane_ksiazki(uzytkownik3, powiesc).

```python
% Fakty
ksiazka('W pustyni i w puszczy', sienkiewicz, przygodowa).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Pan Tadeusz', mickiewicz, poezja).
ksiazka('Zbrodnia i kara', dostojewski, psychologiczna).
ksiazka('Duma i uprzedzenie', austen, romans).
ksiazka('Gra o tron', martin, fantasy).
ksiazka('Metro 2033', glukhovsky, sci_fi).
ksiazka('Harry Potter i Kamień Filozoficzny', rowling, fantasy).
ksiazka('Mały Książę', saint_exupery, filozoficzna).
ksiazka('Lalka', prus, powiesc).

:- dynamic ocena/3.
ocena(uzytkownik1, 'W pustyni i w puszczy', 5).
ocena(uzytkownik1, 'Hobbit', 4).
ocena(uzytkownik1, 'Pan Tadeusz', 3).
ocena(uzytkownik1, 'Gra o tron', 4).
ocena(uzytkownik1, 'Metro 2033', 2).
ocena(uzytkownik2, 'W pustyni i w puszczy', 4).
ocena(uzytkownik2, 'Harry Potter i Kamień Filozoficzny', 5).
ocena(uzytkownik2, 'Mały Książę', 4).
ocena(uzytkownik3, 'Duma i uprzedzenie', 5).
ocena(uzytkownik3, 'Lalka', 4).
ocena(uzytkownik4, 'Metro 2033', 5).
ocena(uzytkownik4, 'Gra o tron', 3).
ocena(uzytkownik4, 'Hobbit', 4).

preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik2, przygodowa).
preferencje(uzytkownik3, romans).
preferencje(uzytkownik4, sci_fi).
preferencje(uzytkownik4, fantasy).

% Reguły

nalezy_do_gatunku(Tytul, Gatunek) :-
    ksiazka(Tytul, _, Gatunek).

polecane_ksiazki(Uzytkownik, Gatunek) :-
    findall(Tytul, (
        ksiazka(Tytul, _, Gatunek),
        ocena(Uzytkownik, Tytul, Ocena),
        Ocena >= 4
    ), Polecane),
    writeln(Polecane).
```
